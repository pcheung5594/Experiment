```python
import win32com
import win32com.client
import cx_Oracle
import os
import pandas as pd
```

```python
file_path = r'C:\Path\To\Your\File'
file_name = r'YourFileName.xlsx'
```

```python
def get_file_name(file_name, path):
    """
    Create name from path and filename
    """
    if path:
        file_name = os.path.join(path, file_name)
    return file_name
```

```python
xl = win32com.client.Dispatch('Excel.Application')
```

```python
wb = xl.Workbooks.Open(get_file_name(file_name, file_path), None, True, None, 'login')
```

```python
wb = xl.Workbooks.Open(get_file_name(file_name, file_path), None)
```

```python
ws = wb.Worksheets('SalesOrders')
```

```python
used = ws.UsedRange
nrows = used.Row + used.Rows.Count - 1
ncols = used.Column + used.Columns.Count - 1
```

```python
content = ws.Range(ws.Cells(1, 1), ws.Cells(nrows, ncols)).Value
```

```python
df = pd.DataFrame(list(content))
headers = df.iloc[0]
df.columns = headers 
df = df[1:]
xl.Quit()
```
